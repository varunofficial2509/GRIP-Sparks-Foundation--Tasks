```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
matchdata = pd.read_csv('matches.csv')
```

```python
matchdata.info()
```

```python
matchdata.head()
```

```python
team_info = pd.DataFrame({'Total Matches' : matchdata.team1.value_counts() + matchdata.team2.value_counts() , 'Toss Won' : matchdata.toss_winner.value_counts(), 'Total Won' : matchdata.winner.value_counts(), 
    'Total Loss' : ((matchdata.team1.value_counts() + matchdata.team2.value_counts()) - matchdata.winner.value_counts())})
```

```python
team_info
```

```python
team_info = team_info.reset_index()
team_info.rename(columns = {'index':'Teams'}, inplace = True)
team_info
```

```python
winloss = team_info['Total Won'] / team_info['Total Matches']
winloss = pd.DataFrame({'Winloss Ratio': team_info['Total Won'] / team_info['Total Matches']})
winloss= winloss.round(2)
team_info = team_info.join(winloss)
team_info
```

```python
team_info
```

```python
team_info.sort_values("Winloss Ratio", axis = 0, ascending = True,inplace = True, na_position ='last')
```

```python
team_info.set_index("Teams", inplace = True)
```

```python
team_info
```

```python
def annotation_plot(ax,w,h):                                    # function to add data to plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for p in ax.patches:
        ax.annotate('{0:.1f}'.format(p.get_height()), (p.get_x()+w, p.get_height()+h))
```

```python
batsmen_stats = playerdata.groupby("batsman").agg({'ball': 'count','batsman_runs': 'sum'})
batsmen_stats.rename(columns={'ball':'balls', 'batsman_runs': 'runs'}, inplace=True)
batsmen_stats = batsmen_stats.sort_values(['balls','runs'], ascending=False)
batsmen_stats.head(15)
```

```python
batsmen_stats['batting_strike_rate'] = batsmen_stats['runs']/batsmen_stats['balls'] * 100
batsmen_stats['batting_strike_rate'] = batsmen_stats['batting_strike_rate'].round(2)
batsmen_stats.head(10)
```

```python
alist = []
for r in playerdata.batsman_runs.unique():
    lookuplist = playerdata[playerdata.batsman_runs == r].groupby('batsman')['batsman'].count()
    batsmen_stats[str(r) + 's'] = trybuild(lookuplist, batsmen_stats)
    try:
        alist.append(lookuplist[r])
    except KeyError:
        alist.append(0)
TopBatsman = batsmen_stats.sort_values(['balls','runs'], ascending=False)[:20]
TopBatsman.head()
```

```python
#Build a dictionary of Matches player by each batsman
played = {}
def BuildPlayedDict(x):
    #print(x.shape, x.shape[0], x.shape[1])
    for p in x.batsman.unique():
        if p in played:
            played[p] += 1
        else:
            played[p] = 1

playerdata.groupby('match_id').apply(BuildPlayedDict)
import operator
```

```python
TopBatsman['matches_played'] = [played[p] for p in TopBatsman.index]
TopBatsman['average']= TopBatsman['runs']/TopBatsman['matches_played']

TopBatsman['6s/match'] = TopBatsman['6s']/TopBatsman['matches_played']  
TopBatsman['6s/match'].median()

TopBatsman['4s/match'] = TopBatsman['4s']/TopBatsman['matches_played']  
TopBatsman['4s/match']
TopBatsman
```

# Bowler stats

```python
bowler_wickets = playerdata.groupby('bowler').aggregate({'ball': 'count', 'total_runs': 'sum', 'player_dismissed' : 'count'})
bowler_wickets.columns = ['runs','balls','wickets']
TopBowlers = bowler_wickets.sort_values(['wickets'], ascending=False)[:15]
TopBowlers
```

```python
TopBowlers['strikerate'] = TopBowlers['runs']/TopBowlers['wickets']
TopBowlers = TopBowlers.sort_values(['strikerate'], ascending=False)[:20]
TopBowlers
```

```python
TopBowlers['strikerate'].plot.bar()
```

#### malinga has less strike rate as shown above in top 15 bowlers

```python
TopBowlers['economy'] = TopBowlers['runs']/(TopBowlers['balls']/6)
TopBowlers = TopBowlers.sort_values(['economy'], ascending=True)[:20]
TopBowlers
```

```python
plt.figure(figsize=(10,8))
plt.bar(np.arange(len(TopBowlers)),TopBowlers['economy'])
plt.xticks(ticks=np.arange(len(TopBowlers)),labels=TopBowlers.index,rotation=90)
plt.xlabel('Bowler')
plt.ylabel('economy')
plt.title('Bowlers Economy')
plt.show()
```

#### Umesh Yadav has less economy rate compared to all other top bowlers
